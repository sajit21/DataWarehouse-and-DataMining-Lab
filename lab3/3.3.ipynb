```python
# Lab 3.3. Divide dataset into 2 clusters using KMeans++ initialization
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = [(2,8),(3,2),(1,4),(4,6),(3,5),(2,3),(5,7),
        (4,8),(4,2),(1,3),(4,5),(3,4),(7,4),(2,1)]
# KMeans++ initialization
km = KMeans(n_clusters=2, init='k-means++')
km.fit(data)
centers = km.cluster_centers_
labels = km.labels_
print("Cluster Centers:", *centers)
print("Cluster Labels:", *labels)
# Splitting points by cluster label
cluster1 = []
cluster2 = []
for i in range(len(labels)):
    if labels[i] == 0:
        cluster1.append(data[i])
    else:
        cluster2.append(data[i])

print("Cluster 1:", cluster1)
print("Cluster 2:", cluster2)
# Visualizing clusters
plt.figure(figsize=(4,4))
# Cluster 1
x1, y1 = zip(*cluster1)
plt.scatter(x1, y1, marker='+', color='red', label='Cluster 1')
# Cluster 2
x2, y2 = zip(*cluster2)
plt.scatter(x2, y2, marker='x', color='green', label='Cluster 2')
# Cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='o', s=50, c='black', label='Centers')
plt.title("KMeans++ (2 Clusters)")
plt.legend()
plt.show()
```
