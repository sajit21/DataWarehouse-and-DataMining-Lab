```python
# Lab 2.2: Demonstrate various data cleaning operations

# Step 1: Import necessary libraries
import pandas as pd
import numpy as np

# Step 2: Load the dataset
file_path = r'C:\Users\Sajit\OneDrive\Desktop\data mining\datasets\employees.csv'
data = pd.read_csv(file_path)

# Step 3: Display original data (first 25 rows)
print("Original Data (First 25 rows):")
print(data.head(25))  # Same as data[0:25]

# ---------------------------------------------
# Step 4: Replacing missing values in Salary with the mean
data_mean_filled = data.copy()
data_mean_filled['Salary'] = data_mean_filled['Salary'].fillna(data_mean_filled['Salary'].mean())

print("\nData After Filling Missing 'Salary' with Mean:")
print(data_mean_filled.head(25))

# ---------------------------------------------
# Step 5: Re-loading original data for interpolation
data_interp = pd.read_csv(file_path)

# Step 6: Replacing missing values in Salary with interpolated values (linear method)
data_interp['Salary'] = data_interp['Salary'].interpolate(method='linear')

# Step 7: Replacing missing 'First Name' with 'Unknown'
data_interp['First Name'] = data_interp['First Name'].fillna("Unknown")

print("\nData After Interpolation and Filling 'First Name':")
print(data_interp.head(25))

```
